```python
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])

df['Item'].isnull().sum()
```

```python
print(df['Price Per Unit'].median())
```

```python
df['Price Per Unit'] = df['Price Per Unit'].fillna(23.0)

df['Price Per Unit'].isnull().sum()
```

```python
print(df['Quantity'].median())
```

```python
df['Quantity'] = df['Quantity'].fillna(6.0)

df['Quantity'].isnull().sum()
```

```python
df['Total Spent'] = df['Price Per Unit'] * df['Quantity']

df['Total Spent'].isnull().sum()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

df.info()
```
